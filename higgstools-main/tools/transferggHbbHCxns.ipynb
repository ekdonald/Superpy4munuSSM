## `gg -> H` + `pp -> bbH` cross sections

```python
# see https://gitlab.com/jonaswittbrodt/ScannerS/-/blob/master/data/info.md for explanations
colnames=[
        'mass',
        'bbh_bbe_pp7',
        'bbh_bbe_pp8',
        'bbh_bbe_pp13',
        'bbh_bbe_pp14',
        'bbh_bbe_ppbar2',
        'bbh_bbo_pp7',
        'bbh_bbo_pp8',
        'bbh_bbo_pp13',
        'bbh_bbo_pp14',
        'bbh_bbo_ppbar2',
        'ggh_bbe_pp7',
        'ggh_bbe_pp8',
        'ggh_bbe_pp13',
        'ggh_bbe_pp14',
        'ggh_bbe_ppbar2',
        'ggh_bbo_pp7',
        'ggh_bbo_pp8',
        'ggh_bbo_pp13',
        'ggh_bbo_pp14',
        'ggh_bbo_ppbar2',
        'ggh_tte_pp7',
        'ggh_tte_pp8',
        'ggh_tte_pp13',
        'ggh_tte_pp14',
        'ggh_tte_ppbar2',
        'ggh_tto_pp7',
        'ggh_tto_pp8',
        'ggh_tto_pp13',
        'ggh_tto_pp14',
        'ggh_tto_ppbar2',
        'ggh_tbe_pp7',
        'ggh_tbe_pp8',
        'ggh_tbe_pp13',
        'ggh_tbe_pp14',
        'ggh_tbe_ppbar2',
        'ggh_tbo_pp7',
        'ggh_tbo_pp8',
        'ggh_tbo_pp13',
        'ggh_tbo_pp14',
        'ggh_tbo_ppbar2',
    ]
df = impl.read_csv_from_web(
    "https://gitlab.com/jonaswittbrodt/ScannerS/-/raw/master/data/ggH_bbH.dat?inline=false",
    sep=r"\s+",
    header=None,
    index_col=None
)
df = df.transpose()
df.columns = colnames
df = df.set_index('mass')
df
```

```python
plt.figure()
plt.plot(df.index.values, df.bbh_bbe_pp7, label="bbh_bbe_pp7")
plt.plot(df.index.values, df.bbh_bbo_pp7, label="bbh_bbo_pp7")
plt.plot(df.index.values, df.ggh_bbe_pp7, label="ggh_bbe_pp7")
plt.plot(df.index.values, df.ggh_bbo_pp7, label="ggh_bbo_pp7")
plt.plot(df.index.values, df.ggh_tte_pp7, label="ggh_tte_pp7")
plt.plot(df.index.values, df.ggh_tto_pp7, label="ggh_tto_pp7")
plt.plot(df.index.values, df.ggh_tbo_pp7, label="ggh_tbo_pp7")
plt.yscale("symlog", linthresh=1e-7)
plt.legend()
plt.show()

plt.figure()
plt.plot(df.index.values, df.bbh_bbe_pp8, label="bbh_bbe_pp8")
plt.plot(df.index.values, df.bbh_bbo_pp8, label="bbh_bbo_pp8")
plt.plot(df.index.values, df.ggh_bbe_pp8, label="ggh_bbe_pp8")
plt.plot(df.index.values, df.ggh_bbo_pp8, label="ggh_bbo_pp8")
plt.plot(df.index.values, df.ggh_tte_pp8, label="ggh_tte_pp8")
plt.plot(df.index.values, df.ggh_tto_pp8, label="ggh_tto_pp8")
plt.plot(df.index.values, df.ggh_tbo_pp8, label="ggh_tbo_pp8")
plt.yscale("symlog", linthresh=1e-7)
plt.legend()
plt.show()

plt.figure()
plt.plot(df.index.values, df.bbh_bbe_pp13, label="bbh_bbe_pp13")
plt.plot(df.index.values, df.bbh_bbo_pp13, label="bbh_bbo_pp13")
plt.plot(df.index.values, df.ggh_bbe_pp13, label="ggh_bbe_pp13")
plt.plot(df.index.values, df.ggh_bbo_pp13, label="ggh_bbo_pp13")
plt.plot(df.index.values, df.ggh_tte_pp13, label="ggh_tte_pp13")
plt.plot(df.index.values, df.ggh_tto_pp13, label="ggh_tto_pp13")
plt.plot(df.index.values, df.ggh_tbo_pp13, label="ggh_tbo_pp13")
plt.yscale("symlog", linthresh=1e-7)
plt.legend()
plt.show()

plt.figure()
plt.plot(df.index.values, df.bbh_bbe_pp14, label="bbh_bbe_pp14")
plt.plot(df.index.values, df.bbh_bbo_pp14, label="bbh_bbo_pp14")
plt.plot(df.index.values, df.ggh_bbe_pp14, label="ggh_bbe_pp14")
plt.plot(df.index.values, df.ggh_bbo_pp14, label="ggh_bbo_pp14")
plt.plot(df.index.values, df.ggh_tte_pp14, label="ggh_tte_pp14")
plt.plot(df.index.values, df.ggh_tto_pp14, label="ggh_tto_pp14")
plt.plot(df.index.values, df.ggh_tbo_pp14, label="ggh_tbo_pp14")
plt.yscale("symlog", linthresh=1e-7)
plt.legend()
plt.show()
```

## ggH XS also applicable for ggH effective decay coupling

```python
plt.figure()
SM_norm_pp13 = df.ggh_tte_pp13 + df.ggh_bbe_pp13 + df.ggh_tbe_pp13
SM_norm_pp8 = df.ggh_tte_pp8 + df.ggh_bbe_pp8 + df.ggh_tbe_pp8
plt.plot(df.index.values, abs(df.ggh_tte_pp13/SM_norm_pp13 - df.ggh_tte_pp8/SM_norm_pp8),   label="ggh_tte_diff",  c='b', ls='-')
plt.plot(df.index.values, abs(df.ggh_tto_pp13/SM_norm_pp13 - df.ggh_tto_pp8/SM_norm_pp8),   label="ggh_tto_diff",  c='r', ls='-')
plt.plot(df.index.values, abs(df.ggh_bbe_pp13/SM_norm_pp13 - df.ggh_bbe_pp8/SM_norm_pp8),   label="ggh_bbe_diff",  c='g', ls='-')
plt.plot(df.index.values, abs(df.ggh_bbo_pp13/SM_norm_pp13 - df.ggh_bbo_pp8/SM_norm_pp8),   label="ggh_bbo_diff",  c='cyan', ls='-')
plt.plot(df.index.values, abs(df.ggh_tbo_pp13/SM_norm_pp13 - df.ggh_tbo_pp8/SM_norm_pp8),   label="ggh_tbo_diff",  c='orange', ls='-')
plt.yscale("log")
plt.legend()
plt.show()
```
